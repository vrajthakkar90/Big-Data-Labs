## Exercise #: Word Count

```python
%%file word_count.py
from mrjob.job import MRJob

class MRWordCount(MRJob):

    def mapper(self, _, line): # map function
        for word in line.split(" "):
            yield word.lower(), 1

    #def combiner(self, word, counts):
    #    yield word, sum(counts)

    def reducer(self, word, counts):
        yield word, sum(counts)

if __name__ == '__main__':
    MRWordCount.run()
```

## Exercise #: `Compute SELECT dno, sum(salary) from employee group by dno;`

```python
%%file empsum.py
from mrjob.job import MRJob
import re

class EmpSum(MRJob):
    def mapper(self, key, line):
      record = re.split(',', line)
      dno = record[2]
      salary = int(record[3])
      yield dno, salary

    def reducer(self, dno, salaries):
        yield dno, sum(salaries)

if __name__ == '__main__':
     EmpSum.run()
```

## Exercise #: `Compute department wise maximum salary among employees from Massachusetts (’MA’).`

```python
%%file empsum.py
from mrjob.job import MRJob
import re

class EmpSum(MRJob):
    def mapper(self, key, line):
      record = re.split(',', line)
      dno = record[2]
      salary = int(record[3])
      if(record[4]=="MA"):
        yield dno, salary

    def reducer(self, dno, salaries):
        yield dno, max(salaries)

if __name__ == '__main__':
     EmpSum.run()
```

## Exercise #: `Compute department wise average salary`

```python
%%file empsum.py
from mrjob.job import MRJob
import re

class EmpSum(MRJob):
    def mapper(self, key, line):
      record = re.split(',', line)
      dno = record[2]
      salary = int(record[3])
      yield dno, (salary,1)
    def reducer(self, dno, salary_count_pair):
        total_salary = 0
        total_count = 0

        for salary, count in salary_count_pair:
            total_salary += salary
            total_count += count

        yield dno, (total_salary/total_count)  # Emit total salary and total count for the department

if __name__ == '__main__':
     EmpSum.run()
```

## Exercise #: `Compute total number of male and female employees for each department.`

```python
%%file empsum.py
from mrjob.job import MRJob
import re

class EmpSum(MRJob):
    def mapper(self, key, line):
      record = re.split(',', line)
      dno = record[2]
      Gender = record[5]
      yield dno, Gender

    def reducer(self, dno, Gender):
        total_female = 0
        total_male = 0
        for gen in Gender:
          if(gen=='F'):
              total_female += 1
          else:
              total_male += 1
        yield dno, (total_male, total_female)  # Emit total salary and total count for the department
if __name__ == '__main__':
     EmpSum.run()
```

## Exercise #: Image Counter from Web Log Access

```python
%%file image_counter.py
from mrjob.job import MRJob

class ImageCounter(MRJob):

	def __init__(self, *args, **kwargs):
		super(ImageCounter, self).__init__(*args, **kwargs)
		self.IMAGES = ["jpeg","jpg","png","jif","jfif","jfi","gif","webp","tiff","tif","psd","raw","arw","cr2","nrw","k25","bmp","dib","heif","heic","ind","indd","indt","jp2","j2k","jpx","jpm","mj2"]
		self.JPGs = ["jpeg","jpg","jif","jfif","jfi"]

	def mapper(self, key, line):
		request = line.split(' ')
		url = request[6]
		method = request[5]
		i = url.find('.')
		if ( i != -1):
			extension = url[i+1:]
			if( extension in self.IMAGES and method[1:] == "GET" ):
				if(extension == "gif" ):
					yield "GIF Images", 1
				elif (extension in self.JPGs):
						yield "JPG Images", 1
				else:
						yield "Other Images", 1

	def reducer(self, key, values):
		yield key, sum(values)

if __name__ == '__main__':
     ImageCounter.run()
```
